Load the weather data from the ***North American Mesoscale Forecast System* (NAM)**. I preprocessed this data (hundreds of gigabytes) and pared it down to the data I believe will be most relevant. The data was obtained for a 6 month winter period, November 2012 to April 2013, dates chosen to coincide with the Utah winter ski season (Wasatch Mountain range, in Salt Lake County).

I obtained the avalanche occurrence data from the ***Utah Avalanche Center* (UAC)**, which was generous enough to provide a crowd-sourced report of avalanches which occurred during the 2012-2013 ski season. The UAC is careful to note that the data may not be complete or even accurate, as not all avalanche occurrences can be verified, and often people make reports a few hours after they see the avalanche and must do it by memory. The users who supply the data do not always mark the same part of the avalanche, so in some cases the crown is marked, in other cases the track is marked, in other cases the runout is marked.

The terrain and elevation data comes from the ***Utah Automated Geographic Reference Center* (AGRC)**, which provides elevation data at 10 meter resolution. I use this to compute slopes and aspects for all points. The crowd sourced avalanche data can then be cross-checked with this data for accuracy. 

```python
# Load NAM weather data
with open('../data/processed/NAMdict2012-13.pkl','r') as f:
    weather_data = pickle.load(f)
    
# Load UAC avalanche occurrence data
with open('../data/processed/UACdict2012-13.pkl','r') as f:
    avalanche_data = pickle.load(f)
```

```python
# run some tests to make sure the data is in good shape
weather_data_shape = weather_data['latitude'].shape
num_days = 181
samples_per_day = 8
num_reported_avalanches = len(avalanche_data['date'])
assert weather_data_shape[0] == num_days * samples_per_day # eight samples per day, 181 days. 
for key in weather_data.keys():
    assert weather_data_shape == weather_data[key].shape
assert num_reported_avalanches == len(avalanche_data['latitude'])
assert len(avalanche_data['longitude']) == len(avalanche_data['latitude'])
```

```python
# a function for accessing avalanche data
def get_avalanche(index):
    return avalanche_data['date'][index],avalanche_data['latitude'][index],avalanche_data['longitude'][index]
```

Variations in when they are reported may be biased by storms and visibility, by high volume snow-sport days (because more are triggered or more are spotted), or may be due to changes in weather or snowpack conditions from day to day which make avalanches more or less likely.

```python
# DEM names
lo_res_dem = 'o11140'
hi_res_dems = ['40111b4','40111b5','40111b6','40111b7','40111b8',
               '40111c4','40111c5','40111c6','40111c7','40111c8',
               '40111d4','40111d5','40111d6','40111d7','40111d8',
               '40111e4','40111e5','40111e6','40111e7','40111e8',
               '40111f4','40111f5','40111f6','40111f7','40111f8',
               '40111g4','40111g5','40111g6','40111g7','40111g8']


# function for determining if a x,y point is within appropriate bounds
def within_extent(minx,miny,maxx,maxy,x,y):
    return minx < x and x < maxx and miny < y and y < maxy

def getDEMdata(filename):
    ds = gdal.Open('../data/GIS/rawdems/' + filename + '.dem',GA_ReadOnly)
    width, height = ds.RasterXSize,ds.RasterYSize
    band = ds.GetRasterBand(1)
    elevationData = band.ReadAsArray(0,0,width, height)
    
    gt = ds.GetGeoTransform()
    originx = gt[0]
    originy = gt[3]
    pixelWidth = gt[1] # w-e pixel res
    pixelHeight = gt[5] # n-s pixel res (negative)
    assert gt[2] == 0, gt[4] == 0
    extent = originx, originy + height * pixelHeight, originx + width * pixelWidth, originy
    resolution = pixelWidth, pixelHeight
    
    return elevationData, extent, resolution

def calc_slopes(elevations,resolution):
    [x_slopes,y_slopes] = np.gradient(elevations,resolution)
    gradients = np.sqrt((x_slopes**2) + (y_slopes[1]**2))
    gradients[gradients>500] = 0
    slopes = np.arctan(gradients)
    return slopes

def calc_aspects(elevations,resolution):
    [x_slopes,y_slopes] = np.gradient(elevations,resolution)
    aspects = np.zeros(elevations.shape)
    aspects[x_slopes > 0] = np.arccos(-y_slopes/np.sqrt(x_slopes**2 + y_slopes**2))[x_slopes > 0]
    aspects[x_slopes < 0] = (2*np.pi - np.arccos(-y_slopes/np.sqrt(x_slopes**2 + y_slopes**2)))[x_slopes < 0]
    return aspects
    
# get high resolution DEMS
elevation_data = []
slope_data = []
aspect_data = []
extents = []
resolutions = []
for dem in hi_res_dems:
    data, extent, resolution = getDEMdata(dem)
    slopes = calc_slopes(data,abs(resolution[0]))
    aspects = calc_aspects(data,abs(resolution[0]))
    elevation_data.append(data)
    slope_data.append(slopes)
    aspect_data.append(aspects)
    extents.append(extent)
    resolutions.append(resolution)

# get the existing coordinate system
lo_res_ds = gdal.Open('../data/GIS/rawdems/'+ lo_res_dem +'.dem',GA_ReadOnly)
old_cs= osr.SpatialReference()
old_cs.ImportFromWkt(lo_res_ds.GetProjectionRef())

# create the new coordinate system
wgs84_wkt = """
GEOGCS["WGS 84",
    DATUM["WGS_1984",
        SPHEROID["WGS 84",6378137,298.257223563,
            AUTHORITY["EPSG","7030"]],
        AUTHORITY["EPSG","6326"]],
    PRIMEM["Greenwich",0,
        AUTHORITY["EPSG","8901"]],
    UNIT["degree",0.01745329251994328,
        AUTHORITY["EPSG","9122"]],
    AUTHORITY["EPSG","4326"]]"""
new_cs = osr.SpatialReference()
new_cs.ImportFromWkt(wgs84_wkt)

# create a transform object to convert between coordinate systems
latlon_to_gtiff = osr.CoordinateTransformation(new_cs,old_cs) 

#get the point to transform, pixel (0,0) in this case
lo_res_width, lo_res_height = lo_res_ds.RasterXSize,ds.RasterYSize
lo_res_gt = lo_res_ds.GetGeoTransform()
assert lo_res_gt[2] == 0, lo_res_gt[4] == 0
lo_res_originx = lo_res_gt[0]
lo_res_originy = lo_res_gt[3]
lo_res_pixelWidth = lo_res_gt[1] # w-e pixel res
lo_res_pixelHeight = lo_res_gt[5] # n-s pixel res (negative)
lo_res_data = lo_res_ds.GetRasterBand(1).ReadAsArray(0,0,lo_res_width,lo_res_height)
lo_res_resolution = lo_res_pixelWidth, lo_res_pixelHeight
lo_res_slopes = calc_slopes(lo_res_data,abs(lo_res_resolution[0]))
lo_res_aspects = calc_aspects(lo_res_data,abs(lo_res_resolution[0]))
lo_res_extent = lo_res_originx, lo_res_originy + lo_res_height * lo_res_pixelHeight, lo_res_originx + lo_res_width * lo_res_pixelWidth, lo_res_originy


# append the low resolution data to use as last resort.
elevation_data.append(lo_res_data)
slope_data.append(lo_res_slopes)
aspect_data.append(lo_res_aspects)
extents.append(lo_res_extent)
resolutions.append(lo_res_resolution)
        
def dataAtLatLon(latitude,longitude):
    x, y, _ = latlon_to_gtiff.TransformPoint(longitude,latitude)
    for elevations, slopes, aspects, extent, resolution in zip(elevation_data,slope_data,aspect_data,extents,resolutions):
        if within_extent(*extent,x=x,y=y):
            i, j = int((x - extent[0])/resolution[0]), int((y - extent[3])/resolution[1])
            if elevations[j,i] >= 0:
                return elevations[j,i], slopes[j,i], aspects[j,i]
    return None, None, None
```

```python
# collect elevation data
elevations = []
slopes = []
aspects = []
for i in range(num_reported_avalanches):
    occurrence_date,lat,lon = get_avalanche(i)
    elevation,slope,aspect = dataAtLatLon(lat,lon)
    if not elevation is None:
        elevations.append(elevation)
        slopes.append(slope)
        aspects.append(aspect)
        #print 'avalance occurred at ', elevation, 'm above sea level at (',lat,', ',lon,') on ', occurrence_date
        
```
